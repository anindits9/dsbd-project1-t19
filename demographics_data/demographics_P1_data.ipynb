```python
import pandas as pd
import altair as alt
```

```python
pop_race_df = pd.read_excel('2015-2019_neighborhood_tables_2021.12.21.xlsm', sheet_name='Race')
```

```python
pop_race_df.columns
```

```python
pop_race_df.drop(['Unnamed: 1','Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11'], axis=1, inplace=True)
pop_race_df.iloc[1,0] = 'neighborhood'
pop_race_df.columns = pop_race_df.iloc[1]
pop_race_df.drop([0,1,2,3], inplace=True)
pop_race_df.reset_index(inplace=True, drop=True)
pop_race_df.dropna(inplace=True)
```

```python
pop_race_df
```

```python
df_add = pd.DataFrame(columns= pop_race_df)
df= pop_race_df[pop_race_df.neighborhood == 'Allston']
df2 = pop_race_df[pop_race_df.neighborhood == 'Brighton']
df_add = pd.concat([df, df2])
df_add.loc[5] = ['Allston / Brighton', df_add['White Alone'].sum(), df_add['Black/African-American'].sum(),
                 df_add['Hispanic'].sum(), df_add['Asian alone'].sum(),  df_add['Other Races'].sum()]
df_add.drop([1,4], inplace=True)
pop_race_df = pd.concat([pop_race_df, df_add])
```

```python
df_add = pd.DataFrame(columns= pop_race_df)
df= pop_race_df[pop_race_df.neighborhood == 'South Boston']
df2 = pop_race_df[pop_race_df.neighborhood == 'South Boston Waterfront']
df_add = pd.concat([df, df2])
df_add.loc[5] = ['South Boston / South Boston Waterfront', df_add['White Alone'].sum(), df_add['Black/African-American'].sum(),
                 df_add['Hispanic'].sum(), df_add['Asian alone'].sum(),  df_add['Other Races'].sum()]
df_add.drop([18,19], inplace=True)
pop_race_df = pd.concat([pop_race_df, df_add])
```

```python
df_add = pd.DataFrame(columns= pop_race_df)
df= pop_race_df[pop_race_df.neighborhood == 'Fenway']
df2 = pop_race_df[pop_race_df.neighborhood == 'Longwood']
df_add = pd.concat([df, df2])
df_add.loc[10] = ['Fenway / Kenmore / Audubon Circle / Longwood', df_add['White Alone'].sum(), df_add['Black/African-American'].sum(),
                 df_add['Hispanic'].sum(), df_add['Asian alone'].sum(),  df_add['Other Races'].sum()]
df_add.drop([9,12], inplace=True)
pop_race_df = pd.concat([pop_race_df, df_add])
```

```python
pop_race_df.drop([9,15,12,19], inplace=True)
```

```python
pop_race_df.reset_index(inplace=True, drop=True)
```

```python
pop_race_df
```

```python
df_merged = pd.read_csv('../merged_311.csv')
```

```python
df_merged.columns
df_merged.drop(['Unnamed: 0'], axis=1, inplace=True)
df_merged.head()
```

```python
grps = df_merged.groupby('ontime')
overdue_df = grps.get_group('OVERDUE')
grps= overdue_df.groupby('neighborhood')
df = pd.DataFrame({'neighborhood': grps.size().index, 'overdue_tickets': grps.size()})
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
pop_race_df= pop_race_df.merge(df, on='neighborhood', how='left')
pop_race_df.sort_values(['overdue_tickets'], ascending=False, inplace=True)
```

```python
pop_race_df
```

```python
melt_df=pd.melt(pop_race_df, id_vars=['neighborhood','overdue_tickets'], value_vars=['White Alone','Black/African-American','Hispanic','Asian alone'
                                                                             ,'Other Races'],var_name='Race', value_name='size')
```

```python
melt_df.dropna(inplace=True)
```

```python
melt_df
```

```python
alt.Chart(melt_df).mark_bar().encode(
    x=alt.X('sum(size)', stack="normalize"),
    y='neighborhood',
    color='Race'
)
```
